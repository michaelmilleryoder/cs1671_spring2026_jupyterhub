# Subset and select parts of DataFrames

```python
subset_columns = ['listing_url', 'name', 'description', 'neighborhood_overview', 'neighbourhood', 'price']
subset = listings[subset_columns].copy()
subset.info()
```

```python
# Check for NaN values
subset['description'].isna().sum()
```

```python
# Convert NaN values to empty strings
subset['description_processed'] = subset['description'].fillna('')
subset[['description', 'description_processed']].head()
```

# Lowercasing
There are many functions available. These are "vectorized", i.e., they are optimized to apply to all cells in a column without iterating through things like for loops.

```python
subset['description_processed'] = subset['description_processed'].str.lower()
subset[['description', 'description_processed']].head()
```

# Custom functions

```python
subset.price
```

```python
def convert_currency(dollar_string):
    if type(dollar_string) == float:
        dollar_float = dollar_string
    else:
        dollar_float = float(dollar_string.replace('$', '').replace(',', ''))
    return dollar_float
```

```python
subset['price_float'] = subset.price.map(convert_currency)
subset[['price', 'price_float']]
```

```python
subset.price_float.dtype
```

```python
subset_under100 = subset.loc[subset['price_float']<100, :]
subset_under100
```
